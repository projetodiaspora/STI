## Colaborao entre Autores

```python
import pandas as pd
import numpy as np
from itertools import combinations

# Filtrar os artigos COM DIASPORA
artigos_com_diaspora = dforiginal[dforiginal['DIASPORA'] == 'COM DIASPORA']

# Criar uma lista de listas de autores para cada artigo
lista_autores = artigos_com_diaspora['AUTHOR_NAME'].dropna().apply(lambda x: x.split(';'))

# Gerar todos os pares de coautoria para cada artigo
pares_coautoria = [combinations(set(autores), 2) for autores in lista_autores]

# Achatar a lista de pares de coautoria e contar as ocorrncias de cada par
pares_flat = [par for sublist in pares_coautoria for par in sublist]
pares_contagem = pd.Series(pares_flat).value_counts()

# Extrair todos os autores nicos
autores_unicos = sorted(set([autor for sublist in lista_autores for autor in sublist]))

# Criar uma matriz de zeros com autores como linhas e colunas
matriz_colaboracao_autores = pd.DataFrame(np.zeros((len(autores_unicos), len(autores_unicos))), index=autores_unicos, columns=autores_unicos)

# Preencher a matriz com as contagens de colaboraes
for (autor1, autor2), contagem in pares_contagem.items():
    if autor1 in matriz_colaboracao_autores.index and autor2 in matriz_colaboracao_autores.columns:
        matriz_colaboracao_autores.loc[autor1, autor2] += contagem
        matriz_colaboracao_autores.loc[autor2, autor1] += contagem  # Assume a colaborao como bidirecional

# Mostrar a matriz de colaborao
print(matriz_colaboracao_autores)
```

```python
import pandas as pd
import numpy as np
from itertools import combinations

# Filtrar os artigos SEM DIASPORA
artigos_sem_diaspora = dforiginal[dforiginal['DIASPORA'] == 'SEM DIASPORA']

# Criar uma lista de listas de assuntos para cada artigo SEM DIASPORA
lista_assuntos = artigos_sem_diaspora['DISPLAY_NAME'].dropna().apply(lambda x: x.split(';'))

# Gerar todos os pares de coocorrncia de assuntos para cada artigo
pares_coocorrencia = [combinations(set(assuntos), 2) for assuntos in lista_assuntos]

# Achatar a lista de pares de coocorrncia e contar as ocorrncias de cada par
pares_flat = [par for sublist in pares_coocorrencia for par in sublist]
pares_contagem = pd.Series(pares_flat).value_counts()

# Extrair todos os assuntos nicos
assuntos_unicos = sorted(set([assunto for sublist in lista_assuntos for assunto in sublist]))

# Criar uma matriz de zeros com assuntos como linhas e colunas
matriz_colaboracao_assuntos2 = pd.DataFrame(np.zeros((len(assuntos_unicos), len(assuntos_unicos))), index=assuntos_unicos, columns=assuntos_unicos)

# Preencher a matriz com as contagens de coocorrncias
for (assunto1, assunto2), contagem in pares_contagem.items():
    if assunto1 in matriz_colaboracao_assuntos2.index and assunto2 in matriz_colaboracao_assuntos2.columns:
        matriz_colaboracao_assuntos2.loc[assunto1, assunto2] += contagem
        matriz_colaboracao_assuntos2.loc[assunto2, assunto1] += contagem  # Assume a coocorrncia como bidirecional

# Mostrar a matriz de colaborao entre assuntos
print(matriz_colaboracao_assuntos2)
```

# IMPORTANTE COM DIASPORA

```python
import pandas as pd
import networkx as nx
from itertools import combinations

# Filtrar os artigos COM DIASPORA
artigos_com_diaspora = dforiginal[dforiginal['DIASPORA'] == 'COM DIASPORA']

# Criar a rede de coautoria
G = nx.Graph()
for autores in artigos_com_diaspora['AUTHOR_NAME'].dropna().apply(lambda x: x.split(';')):
    for autor1, autor2 in combinations(autores, 2):
        G.add_edge(autor1.strip(), autor2.strip())

# Calcular a centralidade de intermediao (betweenness centrality)
centralidade = nx.betweenness_centrality(G)

# Ordenar autores pela centralidade de intermediao
autores_ordenados = sorted(centralidade.items(), key=lambda x: x[1], reverse=True)

# Converter em DataFrame para melhor visualizao
df_centralidade = pd.DataFrame(autores_ordenados, columns=['Autor', 'Centralidade de Intermediao'])

# Mostrar os autores com maior centralidade de intermediao
print(df_centralidade.head(10))
```

```python
import pandas as pd
import networkx as nx
from itertools import combinations

# df_centralidade  o DataFrame dos autores com maior centralidade de intermediao

# Criar um subgrafo apenas com os top autores com maior centralidade de intermediao
top_autores = df_centralidade.head(10)['Autor'].tolist()
subgrafo = G.subgraph(top_autores)

# Criar uma estrutura de dados para mapear cada autor aos seus coautores
coautores = {autor: set() for autor in top_autores}

# Preencher a estrutura com os dados da rede de coautoria COM DIASPORA
for autor in top_autores:
    for coautor in subgrafo.neighbors(autor):
        coautores[autor].add(coautor)

# Converter a estrutura de dados em uma DataFrame para facilitar a visualizao
coautores_df = pd.DataFrame([(autor, sorted(list(coautores_set))) for autor, coautores_set in coautores.items()],
                            columns=['Autor', 'Coautores'])

# Agora coautores_df contm os autores e respectivos coautores
print(coautores_df)
```
